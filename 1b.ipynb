```python
import string, random

def make_key():
    a = list(string.ascii_lowercase); s = a[:]
    random.shuffle(s)
    return dict(zip(a, s))

def encrypt(text, key):
    return ''.join((key[c.lower()].upper() if c.isupper() else key[c.lower()]) if c.isalpha() else c for c in text)

def decrypt(text, key):
    return encrypt(text, {v:k for k,v in key.items()})

key = make_key()
p = input('Enter plain text: ')
enc = encrypt(p, key)
print('Ciphertext:', enc)
print('Decrypted:', decrypt(enc, key))
```
