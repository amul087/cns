```python
!pip install pycryptodome -q
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
import binascii

def enc(k, d):
    c = DES.new(k.encode(), DES.MODE_CBC)
    return binascii.hexlify(c.iv).decode(), binascii.hexlify(c.encrypt(pad(d.encode(),8))).decode()

def dec(k, iv, e):
    return unpad(DES.new(k.encode(), DES.MODE_CBC, binascii.unhexlify(iv)).decrypt(binascii.unhexlify(e)),8).decode()

iv, e = enc('12345678', 'naveen')
print('Encrypted:', e)
print('Decrypted:', dec('12345678', iv, e))
```
