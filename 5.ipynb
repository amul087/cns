```python
import hashlib, ipywidgets as w
from IPython.display import display

txt = w.Text(placeholder='Enter text')
algo = w.Dropdown(options=['SHA-1','SHA-512'])
btn = w.Button(description='Hash', button_style='primary')
out = w.Output()

def run(b):
    h = hashlib.sha1() if algo.value=='SHA-1' else hashlib.sha512()
    h.update(txt.value.encode())
    out.clear_output()
    with out: print(algo.value+':', h.hexdigest())

btn.on_click(run)
display(txt, algo, btn, out)
```
