# GitHub Raw File Fetcher
Paste any `raw.githubusercontent.com` URL below and run the cell. It saves the file directly into your Jupyter folder.

```python
# ── Paste your raw GitHub URL here ──────────────────────────
URL = "https://raw.githubusercontent.com/user/repo/main/file.py"
# ────────────────────────────────────────────────────────────

import requests

r = requests.get(URL)
r.raise_for_status()

filename = URL.split('/')[-1]          # auto-detects filename from URL

with open(filename, 'w', encoding='utf-8') as f:
    f.write(r.text)

print(f'Saved: {filename}  ({len(r.text)} chars)')
```

---
### Fetch multiple URLs at once

```python
# Add as many raw URLs as you need
URLS = [
    "https://raw.githubusercontent.com/user/repo/main/file1.py",
    "https://raw.githubusercontent.com/user/repo/main/file2.ipynb",
    "https://raw.githubusercontent.com/user/repo/main/notes.txt",
]

import requests

for url in URLS:
    r = requests.get(url)
    if r.status_code == 200:
        filename = url.split('/')[-1]
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(r.text)
        print(f'Saved: {filename}')
    else:
        print(f'Failed ({r.status_code}): {url}')
```
