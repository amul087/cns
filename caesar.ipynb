```python
def caesar(text, shift):
    return ''.join(chr((ord(c)-(65 if c.isupper() else 97)+shift)%26+(65 if c.isupper() else 97)) if c.isalpha() else c for c in text)

p = input('Text: '); s = int(input('Shift: '))
enc = caesar(p, s)
print('Encrypted:', enc)
print('Decrypted:', caesar(enc, -s))
```
