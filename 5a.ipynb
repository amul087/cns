```python
import hashlib
import ipywidgets as widgets
from IPython.display import display
def compute_hash(b):
    text = input_box.value
    output_area.clear_output()
    with output_area:
        if not text:
            print("Please enter a valid input string.")
            return
        algo = dropdown.value
        h = hashlib.sha1() if algo == "SHA-1" else hashlib.sha512()
        h.update(text.encode('utf-8'))
        print(f"Step 1 - Bytes     : {text.encode('utf-8').hex()}")
        print(f"Step 2 - Algorithm : {algo} initialized")
        print(f"Step 3 - Intermediate hash: {h.hexdigest()}")
        print(f"Step 4 - Final hash: {h.hexdigest()}")
input_box = widgets.Text(placeholder='Type something...', description='Input:')
dropdown = widgets.Dropdown(options=['SHA-1', 'SHA-512'], value='SHA-512', description='Algorithm:')
btn = widgets.Button(description='Compute', button_style='primary')
output_area = widgets.Output()
btn.on_click(compute_hash)
display(input_box, dropdown, btn, output_area)
```
